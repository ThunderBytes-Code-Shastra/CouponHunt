```python
try:
    from urllib3 import urlopen
except ImportError:
    from urllib.request import urlopen
from lxml import etree

Dict= {"offer":[],"desc":[],"code":[],"tnc":[],"companyname":[],"img":[]};

def grabdata(url='https://www.grabon.in/coupon-codes/55548/'):
  print(url)
  try:
    xpathselector = f"""/html/body/main/div[1]/div[2]/div/div[2]/div/ul"""
    response = urlopen(url)
    htmlparser = etree.HTMLParser()
    tree = etree.parse(response, htmlparser)
    xpath = tree.xpath(xpathselector)
    tnc=''.join(xpath[0].itertext()).strip()
  except:
    xpathselector = f"""/html/body/main/div[1]/div[2]/div/div[2]/div/ol[1]"""  
    response = urlopen(url)
    htmlparser = etree.HTMLParser()
    tree = etree.parse(response, htmlparser)
    xpath = tree.xpath(xpathselector)
    tnc=''.join(xpath[0].itertext()).strip()
    try:
      xpathselector = f"""/html/body/main/div[1]/div[2]/div/div[2]/div/ul[1]"""  
      response = urlopen(url)
      htmlparser = etree.HTMLParser()
      tree = etree.parse(response, htmlparser)
      xpath = tree.xpath(xpathselector)
      tnc=''.join(xpath[0].itertext()).strip()
    except:
      xpathselector = f"""/html/body/main/div[1]/div[2]/div/div[2]/div/ol"""
      response = urlopen(url)
      htmlparser = etree.HTMLParser()
      tree = etree.parse(response, htmlparser)
      xpath = tree.xpath(xpathselector)
      tnc=''.join(xpath[0].itertext()).strip()


  response = urlopen(url)
  htmlparser = etree.HTMLParser()
  tree = etree.parse(response, htmlparser)
  xpath = tree.xpath(xpathselector)
  tnc=''.join(xpath[0].itertext()).strip()
  # print(tnc)
  Dict["tnc"]=(tnc)

  xpathselector = f"""/html/body/main/div[1]/div[2]/div/div[1]/div[1]/div/p[1]/span/text()"""
  xpath = tree.xpath(xpathselector)
  offer=xpath[0]
  # print(offer)
  # Description=''.join(xpath[0].itertext()).strip()
  # print(Description)
  Dict["offer"]= offer

  xpathselector = f"""/html/body/main/div[1]/div[2]/div/div[1]/p"""
  xpath = tree.xpath(xpathselector)
  code=''.join(xpath[0].itertext()).strip()
  # print(code)
  Dict["code"]=code

  xpathselector = f"""/html/body/main/div[1]/div[2]/div/div[1]/div[1]/div/p[2]"""
  xpath = tree.xpath(xpathselector)
  code=''.join(xpath[0].itertext()).strip()
  # print(code)
  Dict["desc"]=code

  xpathselector = f"""/html/body/main/div[1]/div[2]/div/div[1]/div[1]/span/img/@alt"""
  url ='https://www.grabon.in/coupon-codes/110553/'
  response = urlopen(url)
  htmlparser = etree.HTMLParser()
  tree = etree.parse(response, htmlparser)
  xpath = tree.xpath(xpathselector)
  Dict["companyname"]=xpath[0]

  xpathselector = f"""/html/body/main/div[1]/div[2]/div/div[1]/div[1]/span/img/@src"""
  url ='https://www.grabon.in/coupon-codes/110553/'
  response = urlopen(url)
  htmlparser = etree.HTMLParser()
  tree = etree.parse(response, htmlparser)
  xpath = tree.xpath(xpathselector)
  Dict["img"]=xpath[0]
  # print(Dict)
  return Dict
```

```python

categories = [{"tourism":[]},{""}]
cardType = ["Credit & Debit Cards","Credit Cards","Debit Cards"]

def get_card(url="https://www.grabon.in/sbi-card-offers/",bankName="SBI"):
    i = 1
    j = 1
    li = []
    while True:
        xpathselector = f"""/html/body/main/section[2]/div/div[1]/ul[{j}]/li[{i}]/div/@data-cid"""
        i+=1
        response = urlopen(url)
        htmlparser = etree.HTMLParser()
        tree = etree.parse(response, htmlparser)
        xpath = tree.xpath(xpathselector)

        xpathselector_ = f"""/html/body/main/section[2]/div/div[1]/ul[{j}]/li[{i}]/div/div[1]/div[2]/span/div"""
        response = urlopen(url)
        htmlparser = etree.HTMLParser()
        tree = etree.parse(response, htmlparser)
        xpath_ = tree.xpath(xpathselector_)
        if not len(xpath_):
            if(j==3): break 
            j+=1
            i=1
            
        else:
            print(xpath)
            di = grabdata(f"https://www.grabon.in/coupon-codes/{xpath[0]}/")
            di['cardType'] = cardType[j-1]
            di["isDeal"]=0 if (''.join(xpath_[0].itertext()).strip()).find("DEAL")==-1 else 1
            li.append(di)
            print(li,end="\n\n")
    return  li
        
    # print(''.join(xpath[0].itertext()).strip())
```
